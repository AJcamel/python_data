```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('./test.xlsx')

# 得票數 (因為有 3 名候選人)
votes = df['votes']
candidates = df ['Name']

# x 軸文字的座標
x_ticks = [1,2,3,4,5]

# bar 的寬度
width = 0.5

# 繪製長條圖
plt.bar(x_ticks, votes, width)

# 設定數值文字在圖表上
for index in range(len(x_ticks)):
    plt.text(
        x_ticks[index] - 0.1,
        votes[index],
        votes[index]
    )

# 設定圖表標題
plt.title('Camel')
plt.ylabel('Chen')
plt.xlabel('AAA')

# 設定 x 和 y 軸的刻度
plt.xticks(x_ticks, candidates)
plt.yticks(np.arange(0, 11000, step=1000))

plt.show()
```

```python
# 隨機產生 100 位同學的成績
x_scores = list(np.random.randint(0,101,size=100))
# 組距 (將成績 0 到 100 分，分成 10 個組距，回傳 11 個數字)
x_bins = np.arange(0,101,10)
# 繪製直方圖 (會回傳每個組距的數量，以及對應組距)
arr_counts, arr_bins, obj = plt.hist(
    x_scores,
    bins=x_bins,
    edgecolor='black',
    linewidth=1.2
)
# 設定數值文字在圖表上
for i in range(len(arr_counts)):
    plt.text(
        arr_bins[i] + 3,
        arr_counts[i],
        arr_counts[i].astype('int32')
    )
# 設定標題
plt.title('MC')
plt.xticks(x_bins)
plt.ylabel('KFC')
plt.xlabel('Bins')
# 顯示圖表
plt.show()
```

```python
# 匯入套件
import matplotlib.pyplot as plt
import numpy as np

# 設定字型 (optional)
import matplotlib
matplotlib.rc('font', family='Microsoft JhengHei')
```

```python
# 建立資料
x = ['AAAA', 'BBBB', 'CCCCC', 'DDD']
y1 = np.array([10,20,15,30])
y2 = np.array([20,25,15,25])
y3 = np.array([12,15,19,6])
y4 = np.array([10,29,13,19])

# 繪製 stacked bar
plt.bar(x, y1, color='r',edgecolor='black', linewidth=1.2)
plt.bar(x, y2, bottom=y1, color='b',edgecolor='black', linewidth=1.2)
plt.bar(x, y3, bottom=y1+y2, color='y',edgecolor='black', linewidth=1.2)
plt.bar(x, y4, bottom=y1+y2+y3, color='g',edgecolor='black', linewidth=1.2)

# 圖表標題
plt.xlabel('隊伍')
plt.ylabel('分數')
plt.legend(['Round1','Round2','Round3','Round4'])
plt.title('每個隊伍在 4 輪中的分數')
plt.xticks(rotation=45)
# 顯示圖表
plt.show()
```

```python
# 匯入套件
import matplotlib.pyplot as plt

# 設定字型 (optional)
import matplotlib
matplotlib.rc('font', family='Microsoft JhengHei')

y_tick_labels = [
    'Python',
    'Scala',
    'C#',
    'Java',
    'PHP'
]

# 設定類別項目的 x 軸座標
y = [1,2,3,4,5]

# 每個類別項目的值
width = [50,10,13,27,22]

# 繪製 Horizontal Bar
plt.barh(
    y, 
    width, 
    tick_label=y_tick_labels, 
    edgecolor='black', 
    linewidth=2
)

# 設定數值文字在圖表上
for i in range(len(y)):
    # plt.text() 依然使用 x, y text 的順序
    plt.text(width[i]+0.5, y[i], width[i])


# 圖表標題
plt.ylabel('使用次數')
plt.xlabel('程式語言')

plt.show()
```

```python
# 匯入套件
import matplotlib.pyplot as plt

# 設定字型 (optional)
import matplotlib
matplotlib.rc('font', family='Microsoft JhengHei')

# 設定群組資料的值 (第1隊、第2隊)
y_score_1 = [35, 60, 18, 29, 77]
y_score_2 = [28, 33, 93, 44, 32]

# 設定 x 軸的順序，這裡要注意的是，group 的 bar chart，要另外指定 x 軸的位置
BAR_WIDTH = 0.4
x_axis_1 = [1,2,3,4,5]
x_axis_2 = [i + BAR_WIDTH for i in x_axis_1]

# 設定 x 刻度的中心位置
ticks = [i + BAR_WIDTH / 2 for i in x_axis_1]

# 設定 x 標題
tick_labels = ["Round 1", "Round 2", "Round 3", "Round 4", "Round 5"]

# 繪製圖表
plt.bar(
    x_axis_1,
    y_score_1,
    BAR_WIDTH,
    color='y',
    label='第 1 隊'
)
plt.bar(
    x_axis_2,
    y_score_2,
    BAR_WIDTH,
    color='b',
    label='第 2 隊'
)
# 設定數值文字在圖表上
for i in range(len(x_axis_1)):
    plt.text(
        x_axis_1[i],
        y_score_1[i],
        y_score_1[i]
    )
    plt.text(
        x_axis_2[i],
        y_score_2[i],
        y_score_2[i]
    )

# 圖表標題
plt.xlabel('回和數')
plt.ylabel('分數')
plt.title('隊伍分數')
plt.xticks(ticks, tick_labels)
plt.legend(loc='upper left', ncols=2)

plt.show()
```
