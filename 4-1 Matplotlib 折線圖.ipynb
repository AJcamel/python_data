```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# df = pd.read_excel('./temp.rtf')

# y = df['Y']

# x = df['X']

# 準備 y 軸的值
y = [x**2 for x in range(1, 9)]

# 準備 x 軸的值
x = [x for x in range(1, 9)]
```

```python
# 繪製線條 (回傳 matplotlib.lines.Line2D 物件)
plt.plot(y)
# 顯示圖片
plt.show()
```

```python
# 將 x 軸範圍設定成 0 到 8，y 軸設定成 0 到 70
plt.plot(y)

# 設定 x 跟 y 值的刻度
'''
plt.axis([xmin, xmax, ymin, ymax])
'''

plt.axis([0, 8, 0, 70])

# 顯示圖片
plt.show()
```

```python
# 設定線寬為 3
plt.plot(y, linewidth=3)
plt.show()
```

```python
plt.plot(x, y, linewidth=3)
plt.title("Apple", fontsize=24)
plt.xlabel("Cat", fontsize=16)
plt.ylabel("AAA", fontsize=16)
plt.tick_params(axis='both', labelsize=20, colors='r')
plt.show()
```

```python
y_1 = [x**2 for x in range(1,9)]
y_2 = [x*3 for x in range(1,9)]
x = [x for x in range(1,9)]
plt.plot(x, y_1, y_2)
plt.title("Apple", fontsize=24)
plt.xlabel("Cat", fontsize=16)
plt.ylabel("AAA", fontsize=16)
plt.tick_params(axis='both', labelsize=20, colors='r')
plt.show()
```

```python
# 用不同顏色與線條樣式繪製圖表
y_1 = [x**4 for x in range(1,5)]
y_2 = [x**3 for x in range(1,5)]
y_3 = [x**2 for x in range(1,5)]
y_4 = [x*10 for x in range(1,5)]
x = [x for x in range(1,5)]


plt.xticks(x)

# 色彩字元和樣式字元可以一起用
plt.plot(
    x, y_1, '-*',
    x, y_2, '-o',
    x, y_3, '-^',
    x, y_4, '-s',
)
plt.title("Apple", fontsize=24)
plt.xlabel("Cat", fontsize=16)
plt.ylabel("AAA", fontsize=16)
plt.tick_params(axis='both', labelsize=20, colors='r')
plt.show()
```

```python
y_1 = [x**4 for x in range(1,5)]
y_2 = [x**3 for x in range(1,5)]
y_3 = [x**2 for x in range(1,5)]
y_4 = [x*10 for x in range(1,5)]
x = [x for x in range(1,5)]

plt.plot(x, y_1, '-*', label='y_1')
plt.plot(x, y_2, '-o', label='y_2')
plt.plot(x, y_3, '-^', label='y_3')
plt.plot(x, y_4, '-s', label='y_4')

plt.legend(loc='upper right', bbox_to_anchor=(1,0))

plt.title("Apple", fontsize=24)
plt.xlabel("Cat", fontsize=16)
plt.ylabel("AAA", fontsize=16)
plt.tick_params(axis='both', labelsize=20, colors='r')

plt.savefig('./output1.jpg', bbox_inches='tight')

plt.show()
```
